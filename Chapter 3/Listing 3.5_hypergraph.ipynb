```python
import hypernetx as hnx

data = {
    0: ("A","B","G"),
    1: ("A","C","D","E","F"),
    2: ("B","F"),
    3: ("A","B","D","E","F","G")
}
H = hnx.Hypergraph(data)
hnx.draw(H)
```
