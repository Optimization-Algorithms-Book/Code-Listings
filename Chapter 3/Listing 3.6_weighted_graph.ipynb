```python
import networkx as nx
import matplotlib.pyplot as plt

# Create an empty weighted graph
G = nx.Graph()

# Add nodes to the graph (representing devices)
G.add_node("Device1", pos=(0,0))
G.add_node("Device2", pos=(0,2))
G.add_node("Device3", pos=(2,0))
G.add_node("Device4", pos=(2,2))

# Add edges to the graph (representing connections)
G.add_weighted_edges_from([("Device1", "Device2", 45.69),
                           ("Device1", "Device3", 56.34),
                           ("Device2", "Device4", 18.5)]) 
pos = nx.get_node_attributes(G, 'pos')

# Draw the graph
nx.draw_networkx(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)})
plt.show()
```

### Adjust node color, edge label positions and font size

```python
# Draw the graph
nx.draw(G, pos, with_labels=False, node_color='gray')

# Draw node labels
labels = {node: node for node in G.nodes()}
node_labels = nx.draw_networkx_labels(G, pos, labels=labels, font_size=12)

# Adjust label positions
offset = 0.15  # Adjust this value as needed
node_labels["Device1"].set_position((pos["Device1"][0], pos["Device1"][1] - offset))
node_labels["Device3"].set_position((pos["Device3"][0], pos["Device3"][1] - offset))
node_labels["Device2"].set_position((pos["Device2"][0], pos["Device2"][1] + offset))
node_labels["Device4"].set_position((pos["Device4"][0], pos["Device4"][1] + offset))

# Draw edge labels
nx.draw_networkx_edge_labels(G, pos, edge_labels={(u, v): d['weight'] for u, v, d in G.edges(data=True)}, font_size=12)

plt.show()
# plt.savefig('CH03_F06_weighted.png', format='png', bbox_inches='tight', dpi=300)
```
