```python
import numpy as np
import matplotlib.pyplot as plt

def acceptance_probability(delta, temperature):
    if delta < 0:
        return 1.0  # Accept better solution
    return np.exp(-delta / temperature)

# Constants for the first plot
temperature_values_1 = np.linspace(0.1, 10.0, 100)
delta_objective_1 = 0.9
acceptance_probabilities_1 = [acceptance_probability(delta_objective_1, temp) for temp in temperature_values_1]

# Constants for the second plot
temperature_2 = 0.5
delta_values_2 = np.linspace(0.1, 0.9, 100)
acceptance_probabilities_2 = [acceptance_probability(delta, temperature_2) for delta in delta_values_2]

# Plotting in subplots
plt.figure(figsize=(12, 5))

# First subplot
plt.subplot(1, 2, 1)
plt.plot(temperature_values_1, acceptance_probabilities_1, label='Acceptance Probability')
plt.xlabel('Temperature T')
plt.ylabel('Acceptance probability')
plt.title('Temperature vs acceptance probability at f=0.9')
plt.gca().invert_xaxis()
plt.grid(True)

# Second subplot
plt.subplot(1, 2, 2)
plt.plot(delta_values_2, acceptance_probabilities_2, label='Acceptance Probability')
plt.xlabel('Objective function change f')
plt.ylabel('Acceptance probability')
plt.title('Objective function change vs acceptance probability at T=0.5')
plt.grid(True)

plt.tight_layout()
plt.savefig('CH05_F5_Trend.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```
