```python
import numpy as np
import matplotlib.pyplot as plt

def acceptance_probability(delta, temperature):
    if delta < 0:
        return 1.0  # Accept better solution assuming minimization problem
    return np.exp(-delta / temperature)

# Constants for the plot
temperature_values = np.linspace(0.01, 5.0, 20)
delta_values = np.round(np.linspace(0.1, 0.9, 5), 2)  # Round to two decimal places

# Define line styles and markers
line_styles = ['-', '--', '-.', ':', '-']
markers = ['o', 's', '^', 'D', 'x']

# Plotting at different f values with different line styles and markers
plt.figure(figsize=(15, 12))

for i, delta_objective in enumerate(delta_values):
    acceptance_probabilities = [acceptance_probability(delta_objective, temp) for temp in temperature_values]
    plt.plot(temperature_values, acceptance_probabilities, label=f'f={delta_objective}', linestyle=line_styles[i], marker=markers[i])

plt.xlabel('Temperature T', fontsize=15)
plt.ylabel('Acceptance Probability', fontsize=15)
plt.gca().invert_xaxis()
plt.legend(fontsize=14)
plt.grid(True)
plt.savefig('CH05_F5_Trend.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```
