```python
import numpy as np
import matplotlib.pyplot as plt

def exponential_acceptance_probability(T, delta_F):   
    return np.exp(-delta_F/T)
 
def non_exponential_acceptance_probability(T, delta_F, delta_max):
    return np.abs((1-(delta_F/T)/delta_max))

T = np.arange(1, 0.1, -0.05)
delta_F = np.arange(0.1, 1, 0.1)
delta_max=20

X, Y = np.meshgrid(T, delta_F)

fig = plt.figure(figsize = (20,20))

# Exponential acceptance probability
ax = fig.add_subplot(1,2, 1, projection='3d')
p_ex = exponential_acceptance_probability(X, Y)
surf = ax.plot_surface(Y, X, p_ex, cmap=plt.cm.cividis)
plt.xlabel('$\Delta$ Function', fontsize=12, loc="right")
plt.ylabel('Temperature', fontsize=10)
ax.set_zlabel('Acceptance Probability', fontsize=10, rotation = 0)
ax.set_title('Exponential Acceptance Probability', fontsize=13)

# Non-exponential acceptance probability
ax = fig.add_subplot(1,2,2, projection='3d')
p_non = non_exponential_acceptance_probability(X, Y, delta_max)
surf = ax.plot_surface(Y,X, p_non, cmap=plt.cm.cividis)
plt.xlabel('$\Delta$ Function', fontsize=10, loc="right")
plt.ylabel('Temperature', fontsize=10)
ax.set_zlabel('Acceptance Probability', fontsize=10, rotation = 0)
ax.set_title('Non-Exponential Acceptance Probability', fontsize=13)
plt.savefig('acceptance_probability.png', format='png', bbox_inches='tight', dpi=300)
plt.show()
```
