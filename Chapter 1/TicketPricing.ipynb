```python
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -(-20*x*x+6200*x-350000)/1000 # converted into minimization problem
```

```python
res=opt.minimize_scalar(f, method='bounded', bounds=[0, 250])

print("Optimal Ticket Price ($): %.2f" % res.x)
print("Profit f(x) in K$: %.2f" % -res.fun)
```

```python
# plot proft-ticket price curve
plt.figure(figsize=(10, 7))
x = np.linspace(0, 250, 100)
plt.plot(x, -f(x))

# Plot the optima
plt.plot(res.x, -res.fun, "s", color="r")

# Add title and axis names
plt.xlabel("Ticket price ($)", fontsize=20)
plt.ylabel("Profit (K$)", fontsize=20)
plt.show()
# plt.savefig("CH01_F05_Pricing.png", format="png", dpi=300)
```
