```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Acceleration versus EPA Range 
dataset = [[1.7,210], [1.9,270], [2.0,340], [2.2,350], [2.4,355], [2.7,370], [3.2,380], [3.7,410], [4.0,450], [4.5,470], [5.0,500], [5.5,520], [6.0,530]] 
  
df = pd.DataFrame(dataset, columns =['Acceleration time [0-60 mph (s)]', 'EPA Range [miles]']) 
print(df )
```

```python
# scatter plot
sns.set(rc={'figure.figsize':(11.7,8.27)}, font_scale = 1.3)
scatter_plot =sns.scatterplot(x=df['Acceleration time [0-60 mph (s)]'],y=df['EPA Range [miles]'], s=70)
scatter_fig = scatter_plot.get_figure()
scatter_fig.savefig("CH01_F06_EV.png", format="png", dpi=300)
```
