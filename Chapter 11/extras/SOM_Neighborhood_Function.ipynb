# Gaussian Neighborhood Function in SOM

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian(x, sigma):
    return np.exp(-x**2 / (2*sigma**2))

# Create a range of values
x = np.linspace(-3, 3, 100)

# Specify a value for sigma
sigma = 0.61

# Calculate the Gaussian values
y = gaussian(x, sigma)

# Plot the Gaussian function
plt.figure(figsize=(8, 6))
plt.ylim(0)
plt.plot(x, y)
plt.title(f'Gaussian Neighborhood Function ($\sigma$ = {sigma})')  
plt.xlabel('Distance from the winning neuron $LDist(u,v)$', fontsize=14)
plt.ylabel('Influence of the winning neuron $\Theta (u,v)$', fontsize=14)
plt.grid(True)
# Shading the area within one sigma
plt.fill_between(x, y, where=(x >= -sigma) & (x <= sigma), color='lightblue', alpha=0.5)  
# Annotate the borders of the shaded area on the x-axis
plt.text(-sigma, -0.05, f'{round(-sigma, 2)}', ha='center', fontsize=10)
plt.text(sigma, -0.05, f'{round(sigma, 2)}', ha='center', fontsize=10)

plt.show()
```
