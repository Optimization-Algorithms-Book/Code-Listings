## Generate and plot Gaussian distributions with different standard deviations

```python
from numpy import arange
from matplotlib import pyplot
from scipy.stats import norm

# define standard deviations
std_devs = [0.34, 0.52, 0.85]

# x-axis for the plot
x_axis = arange(-5, 5, 0.001)
pyplot.figure(figsize=(8, 6))
# plot low standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[0]), label=r"$\sigma$ = {:.2f}".format(std_devs[0]), linestyle='-')
# plot medium standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[1]), label=r"$\sigma$ = {:.2f}".format(std_devs[1]), linestyle='--')
# plot high standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[2]), label=r"$\sigma$ = {:.2f}".format(std_devs[2]), linestyle=':')
pyplot.grid(color='gray', linestyle='-', linewidth=0.4)
pyplot.xticks(arange(-5, 6, 1))
pyplot.legend(fontsize="10")
pyplot.xlabel('x')
# pyplot.show()
pyplot.savefig('CH08_Gaussians.png', format='png', bbox_inches='tight', dpi=300)
```
