## Generate and plot Gaussian distributions with different standard deviations

```python
from numpy import arange
from scipy.stats import norm
from matplotlib import pyplot

# define standard deviations
std_devs = [0.34, 0.52, 0.85]

# x-axis for the plot
x_axis = arange(-5, 5, 0.001)
pyplot.figure(figsize=(8, 6))
# plot low standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[0]), label=r"$\sigma$ = {:.2f}".format(std_devs[0]), linestyle='-')
# plot medium standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[1]), label=r"$\sigma$ = {:.2f}".format(std_devs[1]), linestyle='--')
# plot high standard deviation
pyplot.plot(x_axis, norm.pdf(x_axis, 0, std_devs[2]), label=r"$\sigma$ = {:.2f}".format(std_devs[2]), linestyle=':')
pyplot.grid(color='gray', linestyle='-', linewidth=0.4)
pyplot.xticks(arange(-5, 6, 1))
pyplot.legend(fontsize="10")
pyplot.xlabel('x')
pyplot.show()
# pyplot.savefig('CH08_Gaussians.png', format='png', bbox_inches='tight', dpi=300)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MaxNLocator

# Parameters for the first Gaussian distribution (small standard deviation)
mean1 = 0
std_dev1 = 2.1

# Parameters for the second Gaussian distribution (large standard deviation)
mean2 = 0
std_dev2 = 6.7

# Generate data points for the first distribution
x1 = np.linspace(mean1 - 4 * std_dev1, mean1 + 4 * std_dev1, 1000)
y1 = (1 / (std_dev1 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x1 - mean1) / std_dev1) ** 2)

# Generate data points for the second distribution
x2 = np.linspace(mean2 - 4 * std_dev2, mean2 + 4 * std_dev2, 1000)
y2 = (1 / (std_dev2 * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x2 - mean2) / std_dev2) ** 2)

# Determine the common horizontal and vertical limits
x_limits = (min(min(x1), min(x2)), max(max(x1), max(x2)))
y_limits = (min(min(y1), min(y2)), max(max(y1), max(y2)))

# Create subplots
fig, axs = plt.subplots(1, 2, figsize=(13, 5))

# Plot the first Gaussian distribution with shaded region between  Std. Dev.
axs[0].plot(x1, y1, label=f'Small Std. Dev.\nMean={mean1}, Std. Dev={std_dev1}')
axs[0].fill_between(x1, 0, y1, where=[(i >= mean1 - std_dev1) and (i <= mean1 + std_dev1) for i in x1], color='gray', alpha=0.3, label=' Std. Dev.')
axs[0].set_xlabel('x')
axs[0].yaxis.set_major_locator(MaxNLocator(nbins=5))  # Reduce the number of y-axis ticks
axs[0].grid(True)

# Plot the second Gaussian distribution with shaded region between  Std. Dev.
axs[1].plot(x2, y2, label=f'Large Std. Dev.\nMean={mean2}, Std. Dev={std_dev2}')
axs[1].fill_between(x2, 0, y2, where=[(i >= mean2 - std_dev2) and (i <= mean2 + std_dev2) for i in x2], color='gray', alpha=0.3, label=' Std. Dev.')
axs[1].set_xlabel('x')
axs[1].yaxis.set_major_locator(MaxNLocator(nbins=5))  # Reduce the number of y-axis ticks
axs[1].grid(True)

# Set the same horizontal and vertical axes scales for both subplots
axs[0].set_xlim(x_limits)
axs[0].set_ylim(y_limits)
axs[1].set_xlim(x_limits)
axs[1].set_ylim(y_limits)

plt.show()
# plt.savefig('CH08_Gaussians.png', format='png', bbox_inches='tight', dpi=300)
```
